# Exercicio 5

```python
nome = input("Digite o nome:")
while len(nome) < 3:
  nome = input("Invalido! o nome tem que ter mais de 3 letras: ")

idade = int(input("Digite sua idade: "))
while idade <0 or idade > 150:
  idade = int(input("Invalido! a idade tem que ser de 0 a 150: "))

salario = float(input("Digite seu salario: "))
while salario < 0:
  salario = float(input("Invalido! o slario tem que ser maior que 0: "))

sexo = input("Digite seu sexo (f/m): ").lower()
while sexo != 'f' and sexo != 'm':
  sexo = input("Invalido! digite f ou m para definir o sexo: ").lower()

ec = input("Digite o estado civil (s/c/v/d): ").lower()
while ec not in ['s','c','v','d']:
  ec = input("Invalido! digite entre s, c, v, d : ").lower()

print("\n Dados validos:")
print(f"nome: {nome}")
print(f"idade: {idade}")
print(f"salario: {salario}")
print(f"sexo: {sexo}")
print(f"estado civil: {ec}")
```

# Exercicio 6


```python
numero = int(input("Digite um numero (inteiro): "))
nprimo = True

if numero <= 1:
  nprimo = False
else:
  for i in range(2, numero):
    if numero % i == 0:
      nprimo = False
      break

if nprimo:
  print("numero primo")
else:
  print("nao é primo")
```
